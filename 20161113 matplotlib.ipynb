```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from random import uniform
num_rows,num_cols=3,2
fig,axes=plt.subplots(num_rows,num_cols,figsize=(8,12))
for i in range(num_rows):
    for j in range(num_cols):
        m,s=uniform(-1,1),uniform(1,2)
        x=norm.rvs(loc=m,scale=s,size=100) #rvs,好像是离散型正态分布函数
        axes[i,j].hist(x,alpha=.6,bins=20) #hist为直方图，x轴为20个随机观测值的范围（bin）
        t=r'$\mu={0:.1f},\quad\sigma={1:.2f}$'.format(m,s) #quad mean a blank
        axes[i,j].set_title(t)
        axes[i,j].set_xticks([-4,0,4])
        axes[i,j].set_yticks([])
plt.show()
```
